# Grabing Data

```python
#opening data frame

step_data = pd.read_csv("steps.csv", sep=";")
survey_data = pd.read_csv('survey.csv')
```

```python
#combining two dataframes and showing the head of new dataframe

df = survey_data.merge(step_data, on = "id")
df.head(5)
```
